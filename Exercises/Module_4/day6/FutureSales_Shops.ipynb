```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
shops = pd.read_csv('shops_eng.csv')
```

```python
shops.head()
```

```python
shops.sort_values(by=['shop_name'], inplace=True)
```

```python
shops
```

```python
shops['shop_name'] = shops.shop_name.str.lstrip() 
```

```python
shops['shop_city'] = shops.shop_name.str.split(' ').str[0] 
```

```python
shops.head()
```

```python
shops['shop_city'] = shops['shop_city'].replace({'St.':'St. Petersburg', '!':'Yakutsk'})
```

```python
shops
```

```python
shops['is_online'] = shops.shop_name.apply(lambda x:1 if ('online' in x.lower()) or ('outbound' in x.lower()) else 0)
```

```python
shops.head()
```

```python
from  geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="User")
shops['latitude'] = shops['shop_city'].apply(lambda x: geolocator.geocode(x).latitude if geolocator.geocode(x) else None)
```

```python
shops['longitude'] = shops['shop_city'].apply(lambda x: geolocator.geocode(x).longitude if geolocator.geocode(x) else None)
```
