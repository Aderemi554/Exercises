```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#Loading data
category = pd.read_csv('DataSet/categories_eng.csv')
item = pd.read_csv('DataSet/items_eng.csv')
```

# Exercise1
**Categorizing items**

```python
#Obtaining the main category by taking the first word in the category_name column
main_cat = category.category_name.str.split(' - ').str[0]
```

```python
#With the same splitting as above, the second word was chosen as the sub category
sub_cat = category.category_name.str.split(' - ').str[1]
```

```python
#Creating new columns for main and sub categories in the main dataframe(category)
category['main_category'] = main_cat
category['sub_category'] = sub_cat
```

```python
category.head(60)
```

```python
#Obtaining sub_sub_category from sub_category by taking the words in parenthesis where available
sub_sub_cat = category.sub_category.str.split('(').str[1].str.split(')').str[0]
```

```python
#Creating column for sub_sub_category
category['sub_sub_category'] = sub_sub_cat
```

```python
category.head(60)
```

# Exercise 2
**Applying CountVectorizer and TfidfVectorizer**

```python
item
```

**CountVectorizer**

```python
#Applying CountVectorizer(1gram) to item_name colum in dataframe item
bow = CountVectorizer()
name_bow = bow.fit_transform(item.item_name)
tokens = bow.get_feature_names()
pd.DataFrame(data=name_bow.toarray(), columns=tokens)
```

```python
#setting min_df to 2
bow = CountVectorizer(min_df=2)
name_bow = bow.fit_transform(item.item_name)
tokens = bow.get_feature_names()
pd.DataFrame(data=name_bow.toarray(), columns=tokens)
```

```python
#setting min_df, max_df parameters
bow = CountVectorizer(min_df=2, max_df=10)
name_bow = bow.fit_transform(item.item_name)
tokens = bow.get_feature_names()
pd.DataFrame(data=name_bow.toarray(), columns=tokens)
```

```python
#Applying CountVectorizer(1gram + 2grams) to item_name colum in dataframe item
bow = CountVectorizer(ngram_range=(1, 2))
name_bow = bow.fit_transform(item.item_name)
tokens = bow.get_feature_names()
pd.DataFrame(data=name_bow.toarray(), columns=tokens)
```

```python
#setting min_df to 2
bow = CountVectorizer(min_df=2, ngram_range=(1, 2))
name_bow = bow.fit_transform(item.item_name)
tokens = bow.get_feature_names()
pd.DataFrame(data=name_bow.toarray(), columns=tokens)
```

```python
#setting min_df, max_df parameters
bow = CountVectorizer(min_df=2, max_df=5, ngram_range=(1, 2))
name_bow = bow.fit_transform(item.item_name)
tokens = bow.get_feature_names()
pd.DataFrame(data=name_bow.toarray(), columns=tokens)
```

**TfidfVectorizer**

```python
#Applying TfidfVectorizer(1gram) to item_name colum in dataframe item
tfidf = TfidfVectorizer()
name_tfidf = tfidf.fit_transform(item.item_name)
tokens = tfidf.get_feature_names()
pd.DataFrame(data=name_tfidf.toarray(),columns=tokens)
```

```python
#Setting min_df parameter
tfidf = TfidfVectorizer(min_df=2)
name_tfidf = tfidf.fit_transform(item.item_name)
tokens = tfidf.get_feature_names()
pd.DataFrame(data=name_tfidf.toarray(),columns=tokens)
```

```python
tfidf = TfidfVectorizer(min_df=0.1)
name_tfidf = tfidf.fit_transform(item.item_name)
tokens = tfidf.get_feature_names()
pd.DataFrame(data=name_tfidf.toarray(),columns=tokens)
```

```python
#Applying TfidfVectorizer(1gram + 2grams) to item_name colum in dataframe item
tfidf = TfidfVectorizer(ngram_range=(1, 2))
name_tfidf = tfidf.fit_transform(item.item_name)
tokens = tfidf.get_feature_names()
pd.DataFrame(data=name_tfidf.toarray(),columns=tokens)
```

```python
#Setting min_df parameter
tfidf = TfidfVectorizer(min_df=2, ngram_range=(1, 2))
name_tfidf = tfidf.fit_transform(item.item_name)
tokens = tfidf.get_feature_names()
pd.DataFrame(data=name_tfidf.toarray(),columns=tokens)
```

```python
#Setting min_df, max_df parameters
tfidf = TfidfVectorizer(min_df=2, max_df=30, ngram_range=(1, 2))
name_tfidf = tfidf.fit_transform(item.item_name)
tokens = tfidf.get_feature_names()
pd.DataFrame(data=name_tfidf.toarray(),columns=tokens)
```

```python
#Setting min_df, max_df parameters
tfidf = TfidfVectorizer(min_df=1, max_df=10, ngram_range=(1, 2))
name_tfidf = tfidf.fit_transform(item.item_name)
tokens = tfidf.get_feature_names()
pd.DataFrame(data=name_tfidf.toarray(),columns=tokens)
```

```python
#Setting min_df, max_df parameters
tfidf = TfidfVectorizer(min_df=1, max_df=1000, ngram_range=(1, 2))
name_tfidf = tfidf.fit_transform(item.item_name)
tokens = tfidf.get_feature_names()
pd.DataFrame(data=name_tfidf.toarray(),columns=tokens)
```
