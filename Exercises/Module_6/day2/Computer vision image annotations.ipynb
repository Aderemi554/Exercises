### To edit an image making a copy of it is always a good idea!

```python
# Load an image
# And make a copy of the original image to avoid manipulating the original image
image = cv2.imread('img/sunset.jpg')
sunset = image.copy()
```

[Hint](# "Numpy .copy function") 

```python
sunset.shape == image.shape
```

## Drawing

### Drawing lines

```python
plt.figure(figsize=(10,5))
plt.imshow(sunset);
```

```python
# Draw some lines on top of the image 
color = 255,255,255
sunset = cv2.line(sunset,(50,200), (200,200), color, 3)

plt.figure(figsize=(10,5))
plt.imshow(sunset);
```

[Hint](# "cv2.line function") 

### Circles

```python
# Draw some circles on top of the image
color = 255,0,0

sunset = cv2.circle(sunset, (150,120), 20, color, 2)

plt.figure(figsize=(10,5))
plt.imshow(sunset);
```

```python
color = 0,0,0

sunset = cv2.circle(sunset, (200,160), 20, color, 2)

plt.figure(figsize=(10,5))
plt.imshow(sunset);
```

```python
color = 255,0,255

sunset = cv2.circle(sunset, (200,100), 20, color, 2)

plt.figure(figsize=(10,5))
plt.imshow(sunset);
```

[Hint](# "cv2.circle") 

### Ellipses

```python
# Draw some elipses on top of the image
color = 0,255,0

sunset = image.copy()
sunset = cv2.ellipse(sunset, (130,200), (65,100), 0,0,360, color, 2)

plt.figure(figsize=(10,5))
plt.imshow(sunset);
```

```python
color = 255,255,0

sunset = cv2.ellipse(sunset, (100,180), (50,90), 90,0,180, color, 2)

plt.figure(figsize=(10,5))
plt.imshow(sunset);
```

[Hint](# "cv2.ellipse") 

### Poligons

```python
# Lets draw some poligons, first declare the points of the poligon
points = np.array([[50,200], [200,200], [125,100]])
#points = points.reshape((-1, 1, 2))
```

[Hint](# "Example: np.array([[10,5],[120,130],[170,120],[250,210]], np.int32)") 

```python
# Draw the poligon on top of the image
color = 100,0,100

sunset = image.copy()
sunset = cv2.polylines(sunset, [points], True, color, 2)

plt.figure(figsize=(10,5))
plt.imshow(sunset);
```

### Add an overlay on top of an image

```python
# Load an image and create a copy of it you can call it overlay
# Create a rectangle with a fill color (-1) on top of the overlay image
# Combine both images setting some opacity on the overlay
# Add some text on top of the overlay and show the image on screen
```

[Hint](# "cv2.addWeighted")  

```python
img = cv2.imread('img/fish.jpg')
fish = img.copy()
fish_overlay = fish.copy()
```

```python
plt.imshow(fish);
```

```python
color = 0,0,0

fish_overlay = cv2.rectangle(fish_overlay, (0,200), (200,309), color, -1)

plt.imshow(fish_overlay);
```

```python
opacity = 0.3

combined = cv2.addWeighted(fish_overlay, opacity, fish, 1-opacity, 0)

plt.imshow(combined);
```

```python
color = 255,0,255
combined_text = cv2.putText(combined, "I am a fish", (100,60), cv2.FONT_HERSHEY_DUPLEX, 2, color, 4)

plt.imshow(combined_text);
```
