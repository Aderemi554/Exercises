# Computer Vision Histograms

### Import statements

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

## Histograms

```python
# Load some images, calculate and plot their color histograms, one color at the time
# try it with different color spaces BGR, HSV, etc
image = cv2.imread('img/day1.jpg')
img = image.copy()
```

```python
plt.imshow(img)
```

```python
hist_b = cv2.calcHist([img], [0], None, [256], [0,256] )
plt.plot(hist_b, color='b')
```

```python
hist_g = cv2.calcHist([img], [1], None, [256], [0,256] )
plt.plot(hist_g, color='g')
```

```python
hist_r = cv2.calcHist([img], [2], None, [256], [0,256] )
plt.plot(hist_r, color='r') 
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
plt.imshow(img)
```

```python
hist_h = cv2.calcHist([img], [0], None, [256], [0,256] )
plt.plot(hist_h, color='b')
```

```python
hist_s = cv2.calcHist([img], [1], None, [256], [0,256] )
plt.plot(hist_s, color='g')
```

```python
hist_v = cv2.calcHist([img], [2], None, [256], [0,256] )
plt.plot(hist_v, color='r')
```

[Hint](# "cv2.calcHist") 


```python
# Now try to plot all 3 channels at the same time, does the histogram give you an idea of what the image looks like?
colors = ['b', 'g', 'r']

for i,color in enumerate (colors):
    hist = cv2.calcHist([img], [i], None, [256], [0,256] )
    plt.plot(hist, color=color)
    plt.xlim(0,256)
plt.show()
```
