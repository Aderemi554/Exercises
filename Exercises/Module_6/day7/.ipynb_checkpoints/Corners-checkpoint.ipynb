```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Corners

```python
image = cv2.imread('img/chess.jpg')
plt.imshow(image);
```

```python
# We will work with the image above
# Convert it to grayscale
img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
img_gray = np.float32(img_gray)
img = image.copy()

# Detect the corners using the cornerHarris
corners = cv2.cornerHarris(img_gray,10,5,0.06)


# Display the image with the detected corners 
img[corners>0.01*corners.max()]=[255,0,0]

plt.imshow(img);
```

```python
img = cv2.imread('img/chess.jpg')

# Convert img to grayscale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Detect the corners using goodFeaturesToTrack
corners = cv2.goodFeaturesToTrack(img_gray, 100, 0.6, 10)

# Draw the corners on top of img and show it on screen (make sure the corners are visible!)
for corner in corners:
    x,y = corner[0]
    x = int(x)
    y = int(y)
    
    cv2.rectangle(img, (x-5, y-5), (x+5, y+5), (255,0,0), -1)

plt.figure(figsize=(10,5))
plt.imshow(img);
```
