```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = cv2.imread('img/people2.jpg');
img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) 
```

```python
# Given the image above, lets detect faces on the image using haar cascades
# First create a CascadeClassifier using the frontalFace classifier -> Haarcascades/haarcascade_frontalface_default.xml
face_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_frontalface_default.xml')

# Convert the image to grayscale
img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Using the classifiers detect all faces on the image
faces = face_classifier.detectMultiScale(img_gray, minNeighbors=4)

# If no faces detected print("No faces found")
if faces is ():
    print("No faces found")

# Otherwise draw a rectangle on each face that has been detected
for (x,y,w,h) in faces:
    cv2.rectangle(img, (x,y), (x+w, y+h), (0,255,0), 2)
plt.imshow(img);
```

```python
image = cv2.imread('img/peeps2.jpeg');
img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

face_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_frontalface_default.xml')
eye_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_eye.xml')

# Convert the image to grayscale
img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Using the classifiers detect all faces and eyes on the image
faces = face_classifier.detectMultiScale(img_gray, minNeighbors=16)
eyes = eye_classifier.detectMultiScale(img_gray)

# If no faces detected print("No faces found")
if eyes is ():
    print("No faces found") 

# Otherwise draw a rectangle on each face and eye that has been detected
for (fx,fy,fw,fh) in faces:
    cv2.rectangle(img, (fx,fy), (fx+fw, fy+fh), (0,255,0), 4)

for (x,y,w,h) in eyes:
    cv2.rectangle(img, (x,y), (x+w, y+h), (0,0,255), 2)

plt.figure(figsize=(20,8))
plt.imshow(img);
```
