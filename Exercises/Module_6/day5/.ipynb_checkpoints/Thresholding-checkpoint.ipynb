# Computer Vision Thresholding

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Have a look at the 2 images below
text = cv2.imread('img/t1.png')
plt.figure(figsize = (12,10))
plt.imshow(text, cmap='gray')
```

```python
text_2 = cv2.imread('img/t2.png')
plt.figure(figsize = (12,10))
plt.imshow(text_2, cmap='gray')
```

```python
# Apply threshold to the images above to make them look better, try to find the best one
text_img = text.copy()

ret,thresh = cv2.threshold(text_img,180,255,cv2.THRESH_BINARY)
text_img[thresh==255] = 255
plt.figure(figsize=(12,10))
plt.imshow(text_img, cmap='gray');
```

```python
text_img = text.copy()
img = cv2.cvtColor(text_img, cv2.COLOR_BGR2GRAY)
im = cv2.bitwise_not(img)

ret,thresh = cv2.threshold(im, 175, 255, cv2.THRESH_BINARY_INV) 

plt.figure(figsize=(12,10))
plt.imshow(thresh, cmap='gray');
```

```python
# Try to apply adaptiveThreshold and see if that helps, use both ADAPTIVE_THRESH_MEAN_C and ADAPTIVE_THRESH_GAUSSIAN_C
text_img = text.copy()
img = cv2.cvtColor(text_img, cv2.COLOR_BGR2GRAY)

thresh = cv2.adaptiveThreshold(img, 255, cv2.THRESH_BINARY, cv2.ADAPTIVE_THRESH_MEAN_C, 13, 13)

plt.figure(figsize=(12,10))
plt.imshow(thresh, cmap='gray');
```

```python
text_img = text.copy()
img = cv2.cvtColor(text_img, cv2.COLOR_BGR2GRAY)
im = cv2.bitwise_not(img)

thresh = cv2.adaptiveThreshold(im, 255, cv2.THRESH_BINARY, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, 3, 0)

plt.figure(figsize=(12,10))
plt.imshow(thresh, cmap='gray');
```

```python
# Now apply Otsu's thresholding to the same images, always make sure to start with a copy of the original image
text_img = text.copy()
img = cv2.cvtColor(text_img, cv2.COLOR_BGR2GRAY)

ret, otsu = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

plt.figure(figsize=(12,10))
plt.imshow(otsu, cmap='gray');
```

```python
plt.figure(figsize=(12,10))
plt.imshow(img, cmap='gray');
```

```python
# Now try to first apply Gaussian Blur to the image and then Otsu thresholding
text_img = text.copy()
img = cv2.cvtColor(text_img, cv2.COLOR_BGR2GRAY)

Gaus_blur = cv2.GaussianBlur(img, (3,3), 0)
```

```python
plt.figure(figsize=(12,10))
plt.imshow(Gaus_blur, cmap='gray');
```

```python
ret, otsu = cv2.threshold(Gaus_blur, 127, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

plt.figure(figsize=(12,10))
plt.imshow(otsu, cmap='gray');
```
