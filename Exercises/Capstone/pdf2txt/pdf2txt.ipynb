## Method 1
### Using PyPDF2
It did not work well because it got rid of white spaces making it difficult to read the txt file. 

```python
import PyPDF2 as pdf 
```

```python
#Creating the function to extract content of pdf file
def pdf2txt(pdffile):
    pdf_file = pdf.PdfFileReader(pdffile)
    
    #creating txt file to contain pdf's content
    with open('test.txt', 'w') as pdf_output:
        for page in range(pdf_file.getNumPages()):
            data = pdf_file.getPage(page).extractText()
            pdf_output.write(data)
            
    #read and clean data
    with open('test.txt', 'r') as content:
        return content.read().replace('\n', ' ')
```

```python
file_content = pdf2txt('new.pdf') 
```

## Method 2
### Using pdfminer
It worked better, txt file is readable but the code is a bit more complex.

```python
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfpage import PDFPage
import io 
```

```python
def pdf2txt(inPDFfile, outTXTFile):
    inFile = open(inPDFfile, 'rb')
    resMgr = PDFResourceManager()
    retData = io.StringIO()
    TxtConverter = TextConverter(resMgr, retData, laparams=LAParams())
    interpreter = PDFPageInterpreter(resMgr, TxtConverter)
    
    #Process each page in pdf file
    for page in PDFPage.get_pages(inFile):
        interpreter.process_page(page)
        
    txt = retData.getvalue()
    
    #save output data to a txt file
    with open(outTXTFile, 'w') as f:
        f.write(txt)
```

```python
#main
inPDFfile = 'test.pdf'
outTXTFile = 'test.txt'
pdf2txt(inPDFfile, outTXTFile)
```
